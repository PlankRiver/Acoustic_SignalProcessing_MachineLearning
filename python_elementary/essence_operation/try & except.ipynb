1. try...except：捕获和处理意外

2. 纠正 default -> 介绍 else 和 finally

在 Python 的 try 结构中，没有 default 这个关键字。您可能想问的是两个与它功能相关的、非常有用的可选子句：else 和 finally。

a) else 子句：当没有意外发生时

else 块中的代码仅在 try 块中没有发生任何异常时才会执行。

这非常有用，它可以让你把“成功后的代码”和“可能会出错的代码”清晰地分离开。

菜谱比喻：只有当你成功做完菜（try 块顺利完成），你才会把它端上桌（else 块执行）。

try:
    # 可能会出错的部分
    user_input = input("请输入一个数字文件名 (例如: 'data.txt'): ")
    f = open(user_input, 'r')

except FileNotFoundError:
    # 错误处理
    print(f"文件 '{user_input}' 不存在。")

else:
    # 只有在 open() 成功时才执行
    print("文件打开成功，正在读取内容...")
    print(f.read())
    f.close() # 在这里关闭文件是安全的，因为我们知道文件已成功打开
b) finally 子句：无论如何都要做的事

finally 块中的代码总是会执行，无论 try 块中是否发生了异常，也无论异常是否被 except 捕获。

它通常用于执行必须的清理操作，比如关闭文件、释放网络连接或数据库连接。

菜谱比喻：无论菜做成功了、做失败了还是中途放弃了，你最终都必须关掉炉火、打扫厨房（finally 块执行）。

        f.close()
3. raise：主动抛出异常

有时候，Python 的内置操作本身没有错，但根据你的程序逻辑，当前的状态是不对的。这时，你可以使用 raise 关键字手动创建并“抛出”一个异常。

菜谱比喻：你检查了一下买来的鸡蛋，发现它们是坏的。打鸡蛋这个动作本身没问题，但你作为厨师，判断出这个食材不能用，于是你主动大喊“鸡蛋坏了！” (raise BadIngredientError)，并停止做菜。

def set_age(age):
    if not isinstance(age, int):
        # 如果类型不是整数，这在逻辑上是错误的
        raise TypeError("年龄必须是一个整数。")
    if age < 0 or age > 120:
        # 如果年龄不在合理范围内，这在逻辑上也是错误的
        raise ValueError("年龄必须在 0 到 120 之间。")
    print(f"年龄已设置为: {age}")

try:
    set_age(25)     # 成功
    set_age("thirty") # 触发 TypeError
    set_age(-5)     # 触发 ValueError
except (TypeError, ValueError) as e:
    print(f"设置年龄失败，原因: {e}")
总结表格
关键字	何时执行？	主要用途
try	总是首先执行。	包裹可能会出错的代码。
except	当 try 块中发生匹配类型的异常时。	处理错误，防止程序崩溃。
else	当 try 块中没有发生任何异常时。	执行“成功路径”下的代码。
finally	永远执行，无论是否发生异常。	执行必须的清理操作（如关闭文件）。
raise	当代码执行到 raise 语句时。	主动、手动地创建一个异常。
